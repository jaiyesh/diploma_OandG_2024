```python
depth = np.arange(1000,6000,500)

porosity = np.linspace(40,10,10)

perm = np.linspace(500,100,10)

reservoir_pressure = np.linspace(3000,4500,10)
```

```python
len(depth)
```

```python
df = pd.DataFrame({"depth":depth,"porosity":porosity,"permeability":perm,"RPressure":reservoir_pressure})
```

```python
df
```

```python
df.index = df["depth"]
```

```python
# funtion for dealing with individual well data

def dataformat(df,wellbore_name):
    """
    Function for getting data of a particular well from the master dataframe

    inputs ->

    df: master dataframe
    wellbore_name: wellbore whsoe data need to be extracted

    return 

    df_well

    """

    df_well = df[df["NPD_WELL_BORE_NAME"]==wellbore_name]

    df_well = df_well.reset_index()

    df_well = df_well.drop("index",axis = 1)

    df_well["DATEPRD"] = pd.to_datetime(df_well["DATEPRD"])

    return df_well 

```

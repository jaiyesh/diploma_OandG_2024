```python
def sigma(z):

    s = 1/(1+np.exp(-z))

    return s
```

```python
def intialize(X):
    n = X.shape[0]

    w = np.random.randn(n,1)
    b = 0

    return w,b 
```

```python
def fpropagate(X,w,b):

    Z = np.dot(w.T, X) + b 

    A = sigma(Z)

    return A
```

```python
def cost(X,y,w,b):

    A = fpropagate(X,w,b)

    L = y*np.log(A) + (1-y)*np.log(1-A)

    m = X.shape[1]

    J = (-1/m)*np.sum(L)

    return J 
```

```python
def gradient(X,y,w,b):
    
    #A ==> (1,m) <== Y
    #dw = dj/dw ==> (2,1)
    #X ==> (2,m)
    #dw ==> (2,m).(m,1)
    
    m = X.shape[1]
    
#     w,b = initialize(X)
    
    A = fpropagate(X,w,b)
    
    #dJ/dw
    dw = (1/m)*np.dot(X, (A - y).T)
    
    #dJ/db
    db = (1/m)*np.sum(A-y)
    
    return dw , db
```

```python
def backprop(X,y,w,b,epoch):
    m = X.shape[1]

    a = 0.05

    for i in range(epoch):
        dw,db = gradient(X,y,w,b)

        w = w -a*dw
        b = b -a*db 

    w_final =w
    b_final = b

    return w_final, b_final 
```

```python
w,b = intialize(X)
```

```python
wf, bf = backprop(X,y,w,b,100000)
```

```python
def predict(X,w=wf,b=bf):
    yp = np.zeros((1,600))

    Z = np.dot(w.T,X)+b 

    A = sigma(Z)

    for i in range(600):

        if A[0,i] >=0.5:
            yp[0,i] = 1
        else:
            yp[0,i] = 0 
    
    return yp
```

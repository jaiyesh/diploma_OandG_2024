## Normal Distribution

#### np.random.normal(loc=mean,scale=std,size)

```python
x = np.random.normal(loc = 25,scale = 5, size =(10000))
y = np.random.normal(loc = 10,scale = 10, size =(10000))
```

```python
sns.distplot(x)
sns.distplot(y)
plt.grid(True)
```

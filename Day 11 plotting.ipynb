### Relative permeability plots

```python
sw = np.linspace(0,1,100)
```

```python
sw
```

```python
krw = sw**3

kro = (1-sw)**2
```

```python
krw
```

```python
kro
```

```python
# step 1: krw vs sw plot
```

```python
plt.plot(sw,krw)
```

```python
# step 2: kro vs sw plot

plt.plot(sw,kro)
```

```python
# step 3: krw and kro both on the same plot

plt.plot(sw,krw)
plt.plot(sw,kro)
```

```python
# step 4: Give x axis label

plt.plot(sw,krw)
plt.plot(sw,kro)

plt.xlabel("Water Saturation (fraction)")
```

```python
# step 5: give y axis label

plt.plot(sw,krw)
plt.plot(sw,kro)

plt.xlabel("Water Saturation (fraction)")

plt.ylabel("Relative Permeability (fraction)")
```

```python
# step 6: grid


plt.plot(sw,krw)
plt.plot(sw,kro)

plt.xlabel("Water Saturation (fraction)")

plt.ylabel("Relative Permeability (fraction)")


plt.grid()
```

```python
# step 7: adding legend

plt.plot(sw,krw, label = "Water Relative Permeability")
plt.plot(sw,kro, label = "Oil Relative Permeability")

plt.xlabel("Water Saturation (fraction)")

plt.ylabel("Relative Permeability (fraction)")

plt.legend()
plt.grid()
```

```python
# step 8: changing the size of the plot

plt.figure(figsize=(8,6))

plt.plot(sw,krw, label = "Water Relative Permeability")
plt.plot(sw,kro, label = "Oil Relative Permeability")

plt.xlabel("Water Saturation (fraction)")

plt.ylabel("Relative Permeability (fraction)")

plt.legend()
plt.grid()
```

```python
# step 9: change the line colour

plt.figure(figsize=(8,6))

plt.plot(sw,krw, label = "Water Relative Permeability",color="darkblue")
plt.plot(sw,kro, label = "Oil Relative Permeability",color="brown")

plt.xlabel("Water Saturation (fraction)")

plt.ylabel("Relative Permeability (fraction)")

plt.legend()
plt.grid()
```

```python
# step 10: changing the line thickness


plt.figure(figsize=(8,6))

plt.plot(sw,krw, label = "Water Relative Permeability",color="darkblue",lw=0.7)
plt.plot(sw,kro, label = "Oil Relative Permeability",color="brown",lw=5)

plt.xlabel("Water Saturation (fraction)")

plt.ylabel("Relative Permeability (fraction)")

plt.legend()
plt.grid()
```

```python
## step 11: adding title to the plot



plt.figure(figsize=(8,6))

plt.plot(sw,krw, label = "Water Relative Permeability",color="darkblue",lw=0.7)
plt.plot(sw,kro, label = "Oil Relative Permeability",color="brown",lw=5)

plt.xlabel("Water Saturation (fraction)")

plt.ylabel("Relative Permeability (fraction)")

plt.legend()
plt.title("Relative Permeability Plots")
plt.grid()
```

```python
## step 12: styles

plt.style.available
```

```python
plt.style.use("default")

plt.figure(figsize=(8,6))

plt.plot(sw,krw, label = "Water Relative Permeability",color="darkblue",lw=0.7)
plt.plot(sw,kro, label = "Oil Relative Permeability",color="brown",lw=5)

plt.xlabel("Water Saturation (fraction)")

plt.ylabel("Relative Permeability (fraction)")

plt.legend()
plt.title("Relative Permeability Plots")
plt.grid()
```
